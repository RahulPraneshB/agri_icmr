```python
import pandas as pd
from sqlalchemy import create_engine, types
```

```python
import mysql.connector
mydb = mysql.connector.connect(
    host="localhost",
    user="root",
    password="0420",
)
print(mydb)
conn = mydb.cursor(buffered=True)
```

```python
engine = create_engine(f"mysql+pymysql://root:0420@localhost/agri_ICMR", echo=False)
```

```python
conn.execute("create database agri_ICMR")
mydb.commit()
```

```python
# switch to the database and create the table (if not exists) using the existing cursor `conn`
conn.execute("USE agri_ICMR")

create_table_sql = """
CREATE TABLE IF NOT EXISTS agriICMR (
    State_Name VARCHAR(255),
    District_Name VARCHAR(255),
    Year INT,
    Season VARCHAR(100),
    Crop VARCHAR(255),
    Area FLOAT,
    Production FLOAT,
    Yield FLOAT
)
"""
conn.execute(create_table_sql)
mydb.commit()
print("Table 'agriICMR' created or already exists in database 'agri_ICMR'.")
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

import plotly.express as px
data = pd.read_csv("D:\\agri_project\\cleaned_data (3).csv")
display(data)
```

```python
data.to_sql(
    name='agridata',
    con=engine,
    if_exists='replace',  # or 'append' if the table already exists
    index=False,
    dtype={
        'State_Name': types.VARCHAR(length=255),
        'District_Name': types.VARCHAR(length=255),
        'Year': types.INTEGER(),
        'Season': types.VARCHAR(length=100),
        'Crop': types.VARCHAR(length=255),
        'Area': types.FLOAT(),
        'Production': types.FLOAT(),
        'Yield': types.FLOAT()
    }
)
```
