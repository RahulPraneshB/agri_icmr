```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
data = pd.read_csv("D:\\agri_project\\ICRISAT-District Level Data - ICRISAT-District Level Data.csv")
display(data)
```

```python
import numpy as np

for col in data.columns:
    if data[col].dtype == 'float64':
        data[col] = data[col].apply(lambda x: 0 if x < 0 else x)

display(data)
```

```python
data.columns = [
    col.replace(' (1000 ha)', '_1000_ha')
       .replace(' (1000 tons)', '_1000_tons')
       .replace(' (Kg per ha)', '_Kg_per_ha')
       .replace(' ', '_') # Replace any remaining spaces not part of the unit transformations
    for col in data.columns
]
display(data.head())
```

```python
display(data.columns)
```

```python
data.to_csv('cleaned_data.csv', index=False)
print("Cleaned data saved to cleaned_data.csv")
```

```python
rice_production_by_state = data.groupby('State_Name')['RICE_PRODUCTION_1000_tons'].sum().sort_values(ascending=False)
top_7_rice_producing_states = rice_production_by_state.head(7)
display(top_7_rice_producing_states)
plt.figure(figsize=(12, 6))
sns.barplot(x=top_7_rice_producing_states.index, y=top_7_rice_producing_states.values, palette='viridis', hue=top_7_rice_producing_states.index, legend=False)
plt.title('Top 7 States by Rice Production (1000 tons)')
plt.xlabel('State Name')
plt.ylabel('Total Rice Production (1000 tons)')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
soybean_production_by_state = data.groupby('State_Name')['SOYABEAN_PRODUCTION_1000_tons'].sum().sort_values(ascending=False)
top_5_soybean_producing_states = soybean_production_by_state.head(5)

soybean_area_by_state = data.groupby('State_Name')['SOYABEAN_AREA_1000_ha'].sum()
soybean_yield_efficiency_by_state = (soybean_production_by_state / soybean_area_by_state).fillna(0)

top_5_soybean_yield_efficiency = soybean_yield_efficiency_by_state.loc[top_5_soybean_producing_states.index]

print("Top 5 States by Soybean Production (1000 tons):")
display(top_5_soybean_producing_states)

print("\nYield Efficiency (Kg per ha) for Top 5 Soybean Producing States:")
display(top_5_soybean_yield_efficiency)
```

```python
# Create a figure and a set of subplots
fig, ax1 = plt.subplots(figsize=(14, 7))

# Plotting soybean production as a bar chart on the first axis
ax1.bar(top_5_soybean_producing_states.index, top_5_soybean_producing_states.values, color='skyblue')
ax1.set_xlabel('State Name')
ax1.set_ylabel('Total Soybean Production (1000 tons)', color='black')
ax1.tick_params(axis='y', labelcolor='black')
ax1.set_title('Soybean Production and Yield Efficiency for Top 5 States')
plt.xticks(rotation=45, ha='right')

# Create a second y-axis for yield efficiency
ax2 = ax1.twinx()
ax2.plot(top_5_soybean_yield_efficiency.index, top_5_soybean_yield_efficiency.values, color='red', marker='o', linestyle='-')
ax2.set_ylabel('Yield Efficiency (Kg per ha)', color='red')
ax2.tick_params(axis='y', labelcolor='red')

plt.tight_layout()
plt.show()
```

```python
crops = ['RICE', 'WHEAT', 'MAIZE']

for crop in crops:
    area_col = f'{crop}_AREA_1000_ha'
    production_col = f'{crop}_PRODUCTION_1000_tons'

    if area_col in data.columns and production_col in data.columns:
        plt.figure(figsize=(10, 6))
        sns.scatterplot(data=data, x=area_col, y=production_col)
        plt.title(f'Impact of Area Cultivated on {crop} Production')
        plt.xlabel(area_col)
        plt.ylabel(production_col)
        plt.grid(True)
        plt.show()
    else:
        print(f"Area or Production column not found for {crop}")
```

```python
# Define crops and their respective columns using the actual column names from the dataframe
crops = {
    "Rice":  ('RICE_AREA_1000_ha',  'RICE_PRODUCTION_1000_tons'),
    "Wheat": ('WHEAT_AREA_1000_ha', 'WHEAT_PRODUCTION_1000_tons'),
    "Maize": ('MAIZE_AREA_1000_ha', 'MAIZE_PRODUCTION_1000_tons'),
}

plt.figure(figsize=(15,5))

for i, (crop, (area_col, prod_col)) in enumerate(crops.items(), 1):
    plt.subplot(1, 3, i)

    # Scatter plot
    plt.scatter(data[area_col], data[prod_col], alpha=0.6, label=f"{crop} Data")

    # Fit linear regression line
    z = np.polyfit(data[area_col].fillna(0), data[prod_col].fillna(0), 1)
    p = np.poly1d(z)
    plt.plot(data[area_col], p(data[area_col]), "r--", label="Trendline")

    plt.title(f"{crop}: Area vs Production")
    plt.xlabel("Area (1000 Ha)")
    plt.ylabel("Production (1000 Tons)")
    plt.legend()

plt.tight_layout()
plt.show()
```

```python
rice_yield_by_state = data.groupby('State_Name')['RICE_YIELD_Kg_per_ha'].mean().sort_values(ascending=False)
wheat_yield_by_state = data.groupby('State_Name')['WHEAT_YIELD_Kg_per_ha'].mean().sort_values(ascending=False)
display(rice_yield_by_state.head(10))
display(wheat_yield_by_state.head(10))
yield_comparison_df = pd.DataFrame({
    'Rice Yield': rice_yield_by_state,
    'Wheat Yield': wheat_yield_by_state
}).dropna()

yield_comparison_df.plot(kind='bar', figsize=(15, 7))
plt.title('Rice vs Wheat Yield Across States (Kg per ha)')
plt.xlabel('State Name')
plt.ylabel('Average Yield (Kg per ha)')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```
